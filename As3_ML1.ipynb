```python
cols_df =  cols_df = df.columns.to_list()
for i in ['label','id','cap-color-rate']:
  cols_df.remove(i)
nominal_columns = cols_df
```

```python
nominal_columns
```

```python
# cols_df =  cols_df = df.columns.to_list()
# for i in ['label','id']:
#   cols_df.remove(i)
# nominal_columns = cols_df
dummy_df = pd.get_dummies(df[nominal_columns], drop_first=True) 
df_with_dummy = pd.concat([df, dummy_df], axis=1)
df_with_dummy = df_with_dummy.drop(nominal_columns, axis=1)
print(df_with_dummy.shape)
```

```python
df_with_dummy.head()
```

```python
dummy_df
```

```python
target_col = 'label'
```

```python
from sklearn.model_selection import train_test_split
df_with_dummy = df_with_dummy.drop(columns=['id'])
feature_col = []
for i in df_with_dummy.columns:
  if i not in['label']:
    feature_col.append(i)
X = df_with_dummy[feature_col]
y = df_with_dummy['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.20, random_state=2020)
```

```python
X_train
```

```python
#Answer 4 
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(
    estimator= RandomForestClassifier(),
    param_grid=dict(
        min_samples_leaf=[2,5,10],
        max_depth = [2,3,6],
        criterion=['gini', 'entropy'],
        n_estimators=[100,200],
        random_state = [2020]
    ),
    scoring='f1_macro',
    cv=5,
    n_jobs=-1 # Parallel
)
```

```python


grid_search.fit(X_train, y_train)
```

```python
grid_search_result = grid_search.cv_results_
pd.DataFrame.from_dict(grid_search_result)
```

```python
#Answer 5
grid_search.best_params_
```

```python
my_model = grid_search.best_estimator_
```

```python
my_model
```

```python
import pickle
filename = 'model.sav'
pickle.dump(my_model, open(filename, 'wb'))
```

```python
open_model = pickle.load(open(filename,'rb'))
open_model
```

```python
predictions = open_model.predict(X_test)
```

```python
#Answer 6
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test, predictions , digits=4))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, predictions, labels=[0, 1])
```
